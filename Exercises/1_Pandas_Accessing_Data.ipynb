# Accessing Data

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset

# load dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[0:10, 'job_title':'salary_hour_avg'] # show first 10 rows and columns from job_title to salary_hour_avg
```

```python
df.loc[:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate') #eliminated all the rows with NaN values in the salary_rate column
```

```python
df.loc[0:10, ['job_title','salary_year_avg']] #show only the job_title and salary_year_avg columns for the first 10 rows using a list
```

```python
us_jobs = df[df['job_country'] == 'United States']
us_jobs.loc[:15, ['job_title','salary_year_avg']] #show only the job_title and salary_year_avg columns for the first 15 rows (regardless of the country) using a list
```

```python
us_jobs = df[df['job_country'] == 'United States']
us_jobs[['job_title','salary_year_avg']].iloc[:15] #show only the job_title and salary_year_avg columns for the first 15 rows (not in US)
```

```python
df_date = df[df['job_posted_date'] > '2023-05-01']
df_date.loc[:, ['job_title','job_location','job_posted_date']] #show only the job_title, job_location, and job_posted_date columns for all rows after May 1st, 2023 using a list 
```
