# Exercise - Analyzing Data Science job demand


```python
from datasets import load_dataset
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df_us = df[df['job_country'] == 'United States'].copy()
df_us.head()
```

```python
df_us['job_posted_month'] = pd.to_datetime(df_us['job_posted_date']).dt.strftime('%B') # Convert job_posted to datetime and extract month
df_us['job_posted_year'] = pd.to_datetime(df_us['job_posted_date']).dt.strftime('%Y') # Convert job_posted to datetime and extract year
df_us.loc[:, ['job_posted_date', 'job_posted_month', 'job_posted_year']] # Check the new columns
```

```python
df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size') # Create a pivot table to count the number of job postings per month and job title
```

```python
df_us[df_us['job_posted_year'] != '2023'] # No job posting outside 2023, so we can drop the year column 
df_us.drop(columns=['job_posted_year'], inplace=True) # Drop the year column
```

```python
df_us_copy = df_us.drop_duplicates(subset='job_posted_month').copy() 
df_us_copy.set_index('job_posted_month', inplace=True) # Set job_posted_month as index
df_us_copy.index.name = 'month' # Set index name
df_us_copy.head(3)
```
