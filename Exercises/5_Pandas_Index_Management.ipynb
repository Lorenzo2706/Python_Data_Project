# Exercise - Analyzing Data Science job demand


```python
from datasets import load_dataset
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df_us = df[df['job_country'] == 'United States'].copy()
df_us.head()
```

```python
df_us['job_posted_month'] = pd.to_datetime(df_us['job_posted_date']).dt.strftime('%B') # Convert job_posted to datetime and extract month
df_us['job_posted_year'] = pd.to_datetime(df_us['job_posted_date']).dt.strftime('%Y') # Convert job_posted to datetime and extract year
df_us.loc[:, ['job_posted_date', 'job_posted_month', 'job_posted_year']] # Check the new columns
```

```python
df_us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size') # Create a pivot table to count the number of job postings per month and job title
```

```python
df_us[df_us['job_posted_year'] != '2023'] # No job posting outside 2023, so we can drop the year column 
df_us.drop(columns=['job_posted_year'], inplace=True) # Drop the year column
```

```python
df_us_pivot.reset_index(inplace=True) # Reset index to get job_posted_month as a column
df_us_pivot['month_number'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_us_pivot.sort_values('month_number', inplace=True)
df_us_pivot.set_index('job_posted_month', inplace=True) 
```

```python
df_us_pivot.drop(columns=['month_number'], inplace=True) 
df_us_pivot.drop(columns=['index'], inplace=True) 
df_us_pivot.drop(columns=['level_0'], inplace=True) 
df_us_pivot
```

```python
df_us_pivot.plot(kind='line', figsize=(12, 6), title='Job Postings by Month and Job Title', marker='o') # too messy
```

```python
top_3_jobs = df_us.groupby('job_title_short').size().nlargest(3) # Get the top 3 job titles by count
top_3_jobs = top_3_jobs.index.tolist() # Convert to list 
```

```python
df_us_pivot[top_3_jobs].plot(kind='line', figsize=(8, 6), xlabel='', ylabel='Number of Postings', title='Top 3 Job Postings by Month')
```
